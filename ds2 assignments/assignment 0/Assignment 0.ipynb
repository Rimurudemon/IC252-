3.1. Print the sum of left diagonal and right diagonal of a square matrix. 

Soln: We can use loops to iterate over the elements of the array and add the elements of the left and right diagonal to get the sum of the left and right diagonal. 

We could have also used np.diag to get the diagonal matrix and then sum the elements of the diagonal matrix to get the sum of the left diagnol and np.fliplr to get the right diagonal and then sum the elements of the right diagonal to get the sum of the right diagonal.

```python
mat= np.array([[1,3,4,5,2],[1,5,2,4,3],[5,2,3,4,1,],[1,4,2,6,9],[4,5,2,1,7]])

def left_diagonal_sum ( mat : np . ndarray ) ->float :
        sum = 0
        for i in range(np.size(mat[0])):
            for j in range(np.size(mat[0])):     
                if i == j:
                    sum+=mat[i][j]

                else:
                    pass
        return sum
        pass
def right_diagonal_sum ( mat : np . ndarray ) ->float :
        sum = 0
        for i in range(np.size(mat[0])):
            for j in range(np.size(mat[0])):     
                if i+j == 4:
                    sum+=mat[i][j]

                else:
                    pass
        return sum
        pass

print (f'Left Diagonal Sum of { mat =} is { left_diagonal_sum (mat)}')
print (f'Right Diagonal Sum of { mat =} is { right_diagonal_sum ( mat )}')
```

Question 5: Create a subplot with two plots side by side. Plot a sine wave in the first subplot. Plot a cosine wave in the
second subplot. Add labels, titles, and a legend to the plots.

Soln: We can use the subplot function to create a subplot with two plots side by side; xlabel and ylabel functions to add labels to the x and y axes; title function to add a title to the plot; np.sin and np.cos functions to create the sine and cosine waves.


Adding a legend to this is not necessary as there are two subplots each having only one plot making it clear which plot is which.



```python
import numpy as np
import matplotlib.pyplot as plt

steps= 0.1
x = np.arange(-5,5,steps)
y1 = np.sin(x)
y2 = np.cos(x)
fig, axs = plt.subplots(1, 2)
axs[0].plot(x, y1, color='red', label='sine wave')
axs[0].set_title('Sine Wave')
axs[0].set_xlabel('values of x')
axs[0].set_ylabel('output of function')
axs[0].grid()
axs[1].plot(x, y2, color='blue')
axs[1].set_title('Cosine Wave')
axs[1].set_xlabel('values of x')
axs[1].set_ylabel('output of function')
axs[1].grid()
plt.show()
```
